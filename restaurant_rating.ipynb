## Hyperparameter Tuning using GridSearchCV

### Description of algorithms: https://surprise.readthedocs.io/en/stable/prediction_algorithms_package.html

#### There is large variance in performance for different fold splits, so I run gridsearch 100 times and select the parameters chosen as 'best' the most times

### Tuning KNN-inspired algorithms

```python
param_grid = {"k":[10,20,40], "min_k":[1,2,4,6,8],"verbose":[False],'random_state':[42]}
classes = (KNNBasic,KNNWithMeans,KNNWithZScore,KNNBaseline)
for model in classes:
    start = datetime.datetime.now()
    print(model)
    best_score = []
    best_params = [] 
    gs = GridSearchCV(model,param_grid, measures=["rmse", "mae"], cv=RepeatedKFold(n_splits=3,n_repeats=100))
    gs.fit(data)
    print("Runtime",str(datetime.datetime.now() - start)[:-3])
    print("RMSE: ",round(gs.best_score["rmse"],3))
    print("Best parameters: ",gs.best_params["rmse"])
```

## Compare model performance

#### Mean RMSE over 1000 3-fold CVs

```python
CVResults = pd.DataFrame(columns = ['Model','RMSE','MAE','Timespan'])

NormalPredictor_model = NormalPredictor()
BaselineOnly_model = BaselineOnly(verbose=False)

KNNBasic_model = KNNBasic(k=10, min_k=8,verbose=False,random_state=42)
KNNBaseline_model = KNNBaseline(k=20, min_k=8,verbose=False,random_state=42)
KNNWithMeans_model = KNNWithMeans(k=20, min_k=8,verbose=False,random_state=42)
KNNWithZScore_model = KNNWithZScore(k=10, min_k=8,verbose=False,random_state=42)

SVD_model = SVD(n_epochs=200, lr_all=0.005, reg_all=0.3, verbose=False,random_state=42)
SVDpp_model = SVDpp(n_epochs=100, lr_all=0.001, reg_all=0.2, verbose=False,random_state=42)
NMF_model = NMF(n_epochs=20,verbose=False,random_state=42)
CoClustering_model = CoClustering(n_epochs=20, n_cltr_u=2, n_cltr_i=2,verbose=False,random_state=42)

classes = [NormalPredictor_model,BaselineOnly_model,KNNBasic_model,KNNBaseline_model,KNNWithMeans_model,KNNWithZScore_model,SVD_model,SVDpp_model,NMF_model,CoClustering_model]
for model in classes:
    start = datetime.datetime.now()
    mean_rmses = []
    mean_maes = []
    #for i in range(1):
    out = cross_validate(model, data, ['rmse', 'mae'], RepeatedKFold(n_splits=3,n_repeats=1000))
    mean_rmse = np.mean(out['test_rmse'])
    mean_mae = np.mean(out['test_mae'])  
    #mean_rmses.append(mean_rmse)
    #mean_maes.append(mean_mae)
    cv_time = str(datetime.datetime.now() - start)[:-3]
    #mean_mean_rmse = np.mean(mean_rmses)
    #mean_mean_mae = np.mean(mean_maes)
    CVResults = CVResults.append({'Model': model, 'RMSE': mean_rmse, 'MAE': mean_mae, 'Timespan': cv_time}, ignore_index=True)

CVResults['Model'] = ['NormalPredictor','BaselineOnly','KNNBasic','KNNBaseline','KNNWithMeans','KNNWithZScore','SVD','SVDpp','NMF','CoClustering']
CVResults
```

### Evaluate SVD predictions on one specified test set

```python
from surprise.model_selection import train_test_split as tts
trainset, testset = tts(data, test_size=0.2, random_state=42)
```

```python
predictions = SVD(n_epochs=200, lr_all=0.005, reg_all=0.3, verbose=False, random_state=42).fit(trainset).test(testset)
ResultCatcher = pd.DataFrame(predictions, columns=['User', 'Restaurant', 'Real_Rating', 'Estimated_Rating', 'details'])
ResultCatcher.drop(['details'], axis=1, inplace=True)
ResultCatcher
```

```python
print('MAE:',round(accuracy.mae(predictions),3))
print('RMSE:',round(accuracy.rmse(predictions),3))
```

```python
ResultComparison = pd.DataFrame({'Count': ResultCatcher.groupby(['Real_Rating']).size(),
        'Avg_Est': ResultCatcher.groupby(['Real_Rating']).mean()['Estimated_Rating']
        }).reset_index()
ResultComparison
```
