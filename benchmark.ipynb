## Compare model performance

#### Mean RMSE over 1000 3-fold CVs

```python
CVResults = pd.DataFrame(columns = ['Model','RMSE','MAE','Timespan'])

class PredictMean(AlgoBase):
    def __init__(self):
        AlgoBase.__init__(self)
    def estimate(self, u, i):
        return 3.237

Naive_model = PredictMean()
NormalPredictor_model = NormalPredictor()
BaselineOnly_model = BaselineOnly(verbose=False)

KNNBasic_ub_model = KNNBasic(k=10, min_k=8,verbose=False,random_state=42,sim_options={'user_based':True})
KNNBaseline_ub_model = KNNBaseline(k=40, min_k=8,verbose=False,random_state=42,sim_options={'user_based':True})
KNNWithMeans_ub_model = KNNWithMeans(k=20, min_k=8,verbose=False,random_state=42,sim_options={'user_based':True})
KNNWithZScore_ub_model = KNNWithZScore(k=10, min_k=8,verbose=False,random_state=42,sim_options={'user_based':True})
KNNBasic_ib_model = KNNBasic(k=10, min_k=3,verbose=False,random_state=42,sim_options={'user_based':False})
KNNBaseline_ib_model = KNNBaseline(k=10, min_k=2,verbose=False,random_state=42,sim_options={'user_based':False})
KNNWithMeans_ib_model = KNNWithMeans(k=10, min_k=2,verbose=False,random_state=42,sim_options={'user_based':False})
KNNWithZScore_ib_model = KNNWithZScore(k=10, min_k=4,verbose=False,random_state=42,sim_options={'user_based':False})

SVD_model = SVD(n_epochs=200, lr_all=0.005, reg_all=0.3, verbose=False,random_state=42,biased=True)
# SVDpp_model = SVDpp(n_factors=100, n_epochs=50, lr_all=0.01, reg_all=0.3, verbose=False,random_state=42)
SVDpp_model = SVDpp(n_factors=100, n_epochs=200, lr_all=0.005, reg_all=0.3, verbose=False,random_state=42)
NMF_model = NMF(n_epochs=20,verbose=False,random_state=42,biased=True)
CoClustering_model = CoClustering(n_epochs=20, n_cltr_u=2, n_cltr_i=2,verbose=False,random_state=42)

classes = [Naive_model,NormalPredictor_model,BaselineOnly_model,KNNBasic_ub_model,KNNBaseline_ub_model,KNNWithMeans_ub_model,KNNWithZScore_ub_model,KNNBasic_ib_model,KNNBaseline_ib_model,KNNWithMeans_ib_model,KNNWithZScore_ib_model,SVD_model,SVDpp_model,NMF_model,CoClustering_model]
for model in classes:
    start = datetime.datetime.now()
    out = cross_validate(model, data, ['rmse', 'mae'], RepeatedKFold(n_splits=3,n_repeats=1000))
    mean_rmse = np.mean(out['test_rmse'])
    mean_mae = np.mean(out['test_mae'])  
    cv_time = str(datetime.datetime.now() - start)[:-3]
    CVResults = CVResults.append({'Model': model, 'RMSE': mean_rmse, 'MAE': mean_mae, 'Timespan': cv_time}, ignore_index=True)

CVResults['Model'] = ['Naive','NormalPredictor','BaselineOnly','UB KNNBasic','UB KNNBaseline','UB KNNWithMeans','UB KNNWithZScore','IB KNNBasic','IB KNNBaseline','IB KNNWithMeans','IB KNNWithZScore','SVD','SVDpp','NMF','CoClustering']
CVResults
```

### Evaluate SVD predictions on one specified test set

```python
from surprise.model_selection import train_test_split as tts
trainset, testset = tts(data, test_size=0.2, random_state=42)
```

```python
predictions = SVD(n_epochs=200, lr_all=0.005, reg_all=0.3, verbose=False, random_state=42).fit(trainset).test(testset)
ResultCatcher = pd.DataFrame(predictions, columns=['User', 'Restaurant', 'Real_Rating', 'Estimated_Rating', 'details'])
ResultCatcher.drop(['details'], axis=1, inplace=True)
ResultCatcher
```

```python
print('MAE:',round(accuracy.mae(predictions),3))
print('RMSE:',round(accuracy.rmse(predictions),3))
```

### Compare with naive estimator (just the mean)

```python
class PredictMean(AlgoBase):
    def __init__(self):
        AlgoBase.__init__(self)
    def estimate(self, u, i):
        return 3.237

predictions = PredictMean().fit(trainset).test(testset)
print('MAE:',round(accuracy.mae(predictions),3))
print('RMSE:',round(accuracy.rmse(predictions),3))
```

#### MSE with SVD is (0.9643/1.2417)^2 = 60.31% of the naive estimator
